```python
for col in ['geo_level_1_id', 'height_percentage']:
    sample = X[col].values.tolist()
    sample_normal = []
    m1 = min(sample)
    m2 = max(sample)
    for x in sample:
        z = (x - m1) / (m2 - m1)
        sample_normal.append(z)
    X[col] = np.array(sample_normal).T
    
```

```python
for col in ['age', 'geo_level_2_id', 'geo_level_3_id']:
    sample = X[col].values.tolist()
    X[col] = np.array([np.log10(x) if x != 0 else x for x in sample]).T
    
X.head()
```
